```python
# License: BSD
# Author: Sasank Chilamkurthy
#transferlearning은 기존 모델을 가져와서 약간만 수정해서 사용하는 것이 트랜스퍼러닝
#기존에는 처음부터 학습해서 성능이 높지 않았음
#트랜스퍼러닝은 굉장히 많은 데이터와 시간을 들여서 학습시킨 모델에서 조금만 바꿔서 사용하는 것
#이제 새 코드에서는 트랜스퍼러닝 사용하면 성능 훨씬 높아짐

from __future__ import print_function, division

import torch
import torch.nn as nn
import torch.optim as optim
from torch.optim import lr_scheduler
import numpy as np
import torchvision
from torchvision import datasets, models, transforms
#파이썬 자료구조를 플로팅해주는 라이브러리
import matplotlib.pyplot as plt
import time
import os
import copy

plt.ion()   # interactive mode
```

```python
from google.colab import drive

#load train data and label
drive.mount('/content/drive')

# Data augmentation and normalization for training
# Just normalization for validation
#기존에는 이미지가 아니라 벡터였음
#이미지로 넘어오면서 이게 생김
#이거는 트레인 데이터에 대한 트랜스폼, 아래는 발리데이션에 대한 트랜스폼
data_transforms = {
    #이건 어규멘테이션을 위한 것
    #랜덤리사이즈드크롭은 원래 224보다 큰 이미지인데(256)에서 224개를 뽑음 ex) 맨위에, 맨아래, 조금 아래쪽 등
    #여백이 32*32가 있는 것중에 크롭이 가능
    #이걸로 어규멘테이션 가능
    #어떤 객체가 있으면 그 객체의 모습을 랜덤으로 크롭함
    #랜덤호리존탈플립은 좌우대칭을 랜덤으로 바꿀 수 있음
    #상하반전은 거의 쓰지 않음, 왜냐하면 이미지에서 상하반전이 되면 아예 다른 믈체가 되기 때문
    #따라서 딥러닝도 좌우반전은 학습 잘 하지만 상하반전은 도움이 되지 않음
    #투텐서는 넘피 어레이 파일을 텐서파일로 바꿈
    #노말리제이션은 0~255사이 값을 0~1사이로 바꿈
    #노말리제이션은 웬만하면 성능을 높힘
    #-0.5~0.5가 성능이 더 높음
    #0~1로 하면 음수를 전혀 사용하지 않기 때문에 데이터의 풍부함이 줄어듦
    #따라서 0~1로 하면 크기의 절댓값만 생각하지만 -0.5~0.5로 하면 절댓값의 크기를 고려하면서도 방향성 feature 특징도 고려할 수 있기 떄문에 -0.5~0.5로 노멀리제이션 함
    'train': transforms.Compose([
        transforms.RandomResizedCrop(224),
        transforms.RandomHorizontalFlip(),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ]),
    #여기는 랜덤리사이즈드크롭이 센터크롭으로 바뀌고 랜덤호리존탈필드가 없어짐
    #밸리데이션에서는 항상 동일한 가운데 위치한 플립을 하지 않은 좌우반전을 하지 않는 값을 씀
    'val': transforms.Compose([
        transforms.Resize(256),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ]),
}

data_dir = '/content/drive/MyDrive/인공지능 수업/hymenoptera_data'
#이미지 파일인데 저장해둠
image_datasets = {x: datasets.ImageFolder(os.path.join(data_dir, x),
                                          data_transforms[x])
                  for x in ['train', 'val']}
#데이터로더 클래스에 데이터셋 넣어서 정의함
#중괄호는 for문, train한번, validation한번씩 x에 train, test데이터가 만들어짐
#데이터로더에 트레인 테스트 데이터 다 있는거임
dataloaders = {x: torch.utils.data.DataLoader(image_datasets[x], batch_size=4,
                                             shuffle=True, num_workers=4)
              for x in ['train', 'val']}
dataset_sizes = {x: len(image_datasets[x]) for x in ['train', 'val']}
#숫자가 가리키는 것이 어떤 것인지
class_names = image_datasets['train'].classes

#쿠다 사용, 런타임이 None으로 설정되어 있을때는 GPU 사용해야함
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
```

```python
#이미지 보여주는 것
def imshow(inp, title=None):
    """Imshow for Tensor."""
    inp = inp.numpy().transpose((1, 2, 0))
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    #노말리제이션을 위한 것
    #픽셀 인텐시티가 0~255사이인데 plt에서 imshow하기 위해서는 0~1사이가 되어야 함
    #0~255사이에 있는 것을 노말리제이션 해야함
    #노말리제이션 안하면 이상하게 보임
    inp = std * inp + mean
    inp = np.clip(inp, 0, 1)
    plt.imshow(inp)
    if title is not None:
        plt.title(title)
    plt.pause(0.001)  # pause a bit so that plots are updated


# Get a batch of training data
#데이터로더가 이터레이션 돌음
#데이터로더 포문 돌면 인풋 데이터와 그에 해당하는 레이블 가져올 수 있음
#배치 사이즈가 4이므로 이터레이션 한번에 사진 4개 가져옴
inputs, classes = next(iter(dataloaders['train']))

# Make a grid from batch
out = torchvision.utils.make_grid(inputs)

imshow(out, title=[class_names[x] for x in classes])
```

```python
#크라이테리온은척도, 딥러닝 모델이 얼마나 잘 학습되고 있는지 평가 척도
#옵티마이저는 산을 어떻게 내려갈 것인지 결정
#스케줄러는 러닝 레이트를 설정해 주는것, 밑에 스케줄러.스텝()있음, 한번 스텝마다 러닝 레이트를 조금씩 줄이는 것, 러닝레이트의 스케줄러
#1에폭은 트레인 데이터 전체를 1번 학습하는 것
#25에폭은 모든 학습 데이터 25번을 학습시키는 것
def train_model(model, criterion, optimizer, scheduler, num_epochs=25):
    since = time.time()

    best_model_wts = copy.deepcopy(model.state_dict())
    best_acc = 0.0

    for epoch in range(num_epochs):
        print('Epoch {}/{}'.format(epoch, num_epochs - 1))
        print('-' * 10)

        # Each epoch has a training and validation phase
        #트레인 페이즈, 발리데이션 페이즈 두 개가 있음
        for phase in ['train', 'val']:
            if phase == 'train':
                #학습하는 것
                model.train()  # Set model to training mode
            else:
                #평가하는 것
                model.eval()   # Set model to evaluate mode

            running_loss = 0.0
            running_corrects = 0

            # Iterate over data.
            #한 에폭에서 일어나는 일
            #한 에폭에서 트레인 한번, 발리데이션 한번 페이즈가 발생함
            for inputs, labels in dataloaders[phase]:
                #데이터인 인풋과 정답인 레이블을 가져옴
                inputs = inputs.to(device)
                labels = labels.to(device)

                # zero the parameter gradients
                #옵티마이저에서 지금까지 얼마만큼 내려왔는지의 세팅을 0으로 바꿔줌, 현재 상태에서 생각할 수 있게 함
                optimizer.zero_grad()

                # forward
                # track history if only in train
                #아웃풋은 확률값
                #preds는 예측값
                with torch.set_grad_enabled(phase == 'train'):
                    outputs = model(inputs)
                    _, preds = torch.max(outputs, 1)
                    loss = criterion(outputs, labels)

                    # backward + optimize only if in training phase
                    #트레인이 아니면 백워드 안함
                    if phase == 'train':
                        loss.backward()
                        optimizer.step()

                # statistics
                running_loss += loss.item() * inputs.size(0)
                #얼마나 정확한지 갯수를 세서 러닝 코렉츠에 더함
                running_corrects += torch.sum(preds == labels.data)
            if phase == 'train':
                scheduler.step()

            #로스 계산, 어큐러시 계산
            #로스는 현재 얼마나 딥러닝 모델이 학습되어있는지 척도, 단위가 없음
            #어큐러시는 퍼센트로 단위가 있음
            #서로 다른 모델이나 서로 다른 세팅에서 로스만으로는 모델을 비교할 수 없으
            #로스는 절대적 숫자가 중요하지 않고 처음 랜덤 세팅된 로스에서 얼마나 줄어드는지가 중요함
            #로스는 줄어드는 경향성을 확인하는 것이 중요
            epoch_loss = running_loss / dataset_sizes[phase]
            epoch_acc = running_corrects.double() / dataset_sizes[phase]

            print('{} Loss: {:.4f} Acc: {:.4f}'.format(
                phase, epoch_loss, epoch_acc))

            # deep copy the model
            #베스트 어큐러시도 계산함
            #25번 에폭중에 가장 성능이 높은 것 고름
            #베스트 모델의 웨이트 저장
            if phase == 'val' and epoch_acc > best_acc:
                best_acc = epoch_acc
                #deepcopy는 모델의 메모리 어드레스가 아니라 메모리 전체의 내용을 카피함
                #이걸 앙상블하거나 그냥 사용하기도 가능
                best_model_wts = copy.deepcopy(model.state_dict())

        print()
    #시간 얼마나 걸렸는지 저장함
    time_elapsed = time.time() - since
    print('Training complete in {:.0f}m {:.0f}s'.format(
        time_elapsed // 60, time_elapsed % 60))
    print('Best val Acc: {:4f}'.format(best_acc))

    # load best model weights
    #베스트 모델의 웨이트들을 리턴함
    model.load_state_dict(best_model_wts)
    return model
```

```python
#위에까지는 지금까지 우리가 했던 내용들과 비슷함
#다른 점은 로우 데이터를 직접 학습시키는 것
#우리가 학습했던 모델을 시각화 함
#우리가 학습한 모델에 어떤 데이터를 인풋으로 넣어서 어떻게 예측하는지 시각화함
def visualize_model(model, num_images=6):
    was_training = model.training
    model.eval()
    images_so_far = 0
    fig = plt.figure()

    with torch.no_grad():
        for i, (inputs, labels) in enumerate(dataloaders['val']):
            inputs = inputs.to(device)
            labels = labels.to(device)

            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)

            for j in range(inputs.size()[0]):
                images_so_far += 1
                ax = plt.subplot(num_images//2, 2, images_so_far)
                ax.axis('off')
                ax.set_title('predicted: {}'.format(class_names[preds[j]]))
                #임쇼이미지를 사용해 퍼센트가 아니라 어떤 이미지를 어떻게 예측했다를 눈으로 보여줌
                imshow(inputs.cpu().data[j])

                if images_so_far == num_images:
                    model.train(mode=was_training)
                    return
        model.train(mode=was_training)
```

```python
#여기서부터 트랜스퍼러닝
#기존에 만든 딥러닝 모델을 가져옴
#많은 시간과 데이터를 들여 학습시킨 모델
#이렇게 안하면 개인마다 많은 데이터와 시간을 사용해 학습시킬 수 없음
#따라서 누군가가 만든 딥러닝 모델의 파라미터를 가져와서 우리가 사용하게 살짝 수정해서 사용함
#이런 딥러닝 모델이 일반적으로 성능이 좋은 모델임
#일반적이라는 건 어디서든 잘된다는 것
#나무or개미or돌을 분류하는 것에서도 다 일반적으로 성능이 좋음
#레스넷18은 레이어가 18개 있음
#프레트레인드=트루로 호출하면 미리 학습을 해둔 레스넷18을 사용하는 것을 나타냄
#ft는 fine tuning
#이미 학습된 레스넷18을 다운로드함
#그걸 살짝 수정할거임
#제일 마지막 fc레이어를 좀 바꿀것임
#우리가 예측하고자 하는 클래스 레이블 개수만큼 확률값을 얻도록 바꿔야 한다
model_ft = models.resnet18(pretrained=True)
num_ftrs = model_ft.fc.in_features
# Here the size of each output sample is set to 2.
# Alternatively, it can be generalized to nn.Linear(num_ftrs, len(class_names)).
#우리가 예측해야 하는 레이블 개수가 2개임,벌이랑 개미를 구분해야함
#기존에는 1000개의 오브젝트를 예측하는 모델이었는데 마지막 1000개 예측하는 것을 2개로 바꿈
#.fc가 마지막이고 이걸 2개 예측하게 바꿔야함
#fc이전 레이어들은 기존 웨이트 그대로 쓰고 마지막만 바꿈
#이게 fine tuning(미세 조정)

print(model_ft)
model_ft.fc = nn.Linear(num_ftrs, 2)
print(model_ft)

#쿠다 디바이스로 옮김
model_ft = model_ft.to(device)

criterion = nn.CrossEntropyLoss()

# Observe that all parameters are being optimized
optimizer_ft = optim.SGD(model_ft.parameters(), lr=0.001, momentum=0.9)

# Decay LR by a factor of 0.1 every 7 epochs
#러닝레이트 스케줄러, 러닝 레이트가 10분의 1씩 작아질거임, 7에폭마다
#제일 처음 7개는 0.001, 8~14번쨰는 0.0001 등
exp_lr_scheduler = lr_scheduler.StepLR(optimizer_ft, step_size=7, gamma=0.1)
```

```python
model_ft = train_model(model_ft, criterion, optimizer_ft, exp_lr_scheduler,
                       num_epochs=25)
```

```python
visualize_model(model_ft)
```

```python
#결과적으로 프레딕션한 결과를 비주얼라이제이션 펑션 사용해 결과를 볼 수 있음
#어떻게 예측했는지 문자열로 보여줌
#마지막 fc레이어만 업데이트 할거라는 의미
#앞에서는 모든 레이어의 웨이트를 학습시키지만 여기서는 기존웨이트 그대로 쓰고 마지막 fc만 바꾸겠다는 위미
model_conv = torchvision.models.resnet18(pretrained=True)
for param in model_conv.parameters():
    param.requires_grad = False

# Parameters of newly constructed modules have requires_grad=True by default
num_ftrs = model_conv.fc.in_features
model_conv.fc = nn.Linear(num_ftrs, 2)

#재학습
#여기에는 모델 finetuning하는 거랑 다른점은 위에 for문
model_conv = model_conv.to(device)

criterion = nn.CrossEntropyLoss()

# Observe that only parameters of final layer are being optimized as
# opposed to before.
optimizer_conv = optim.SGD(model_conv.fc.parameters(), lr=0.001, momentum=0.9)

# Decay LR by a factor of 0.1 every 7 epochs
exp_lr_scheduler = lr_scheduler.StepLR(optimizer_conv, step_size=7, gamma=0.1)
```

```python
model_conv = train_model(model_conv, criterion, optimizer_conv,
                         exp_lr_scheduler, num_epochs=25)
```

```python
visualize_model(model_conv)
print("구분")
visualize_model(model_ft)
```
